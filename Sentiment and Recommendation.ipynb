```python
from textblob import TextBlob
import pandas as pd
df=pd.read_csv("F:\cloths-rating.csv")
text=df['Text']
def sentiment_calc(text):
    try:
        return TextBlob(str(text)).sentiment.polarity
    except:
        return None
df['sentiment']=df['Text'].apply(sentiment_calc)
df
```

```python
 df['updated_score']= df['Rating']*df['sentiment']
 df
```

```python
dict1= {'1':[-5,-4,-3.75,-2.5,-2.25,-2,-1.5,-1.25,-1,-0.75,-0.5,-0.25],'2':[-0.24,0.25,0.5,0.75,1],
        '3':[1.01,1.25,1.5,2],'4':[2.01,2.25,2.5,3],'5':[3.01,3.25,3.5,4]}
def classify_score(score):
    if score >=-5 and score <= -0.75:
        return 1
    if score >-0.75 and score <0:
        return 2
    if score >=0 and score < 2:
        return 3
    if score >=2 and score <3.5:
        return 4
    if score >=3.5 and score <=5:
        return 5

df['newscore']=df['updated_score'].apply(classify_score)
df
```

```python
df_pivot=df.pivot_table(index='ProductID', columns='UserID',values='newscore').fillna(0)
df_pivot
```

```python
from scipy.sparse import csr_matrix

df_pivot_matrix= csr_matrix(df_pivot.values)
print(df_pivot_matrix)
```

```python
from textblob import TextBlob
import pandas as pd
from textblob.tokenizers import word_tokenize
from nltk.corpus import stopwords
from textblob.en import polarity
stop_words = set(stopwords.words('english'))
def find_polarity(review):
    global stop_words
    
    blob=TextBlob(review)
    words=blob.words
    sentence=[word for word in words if word not in stop_words]
    sent=" ".join(sentence)
    blob2= TextBlob(sent)
    
    return blob2.sentiment.polarity

data= pd.read_csv("F:\cloths-rating.csv")

apply_functions={'Rating': 'mean', 'Text' : lambda x: ".".join(x)}

new_data= data.groupby(['ProductID']).aggregate(apply_functions)

new_data['polarity']= new_data['Text'].apply(find_polarity)
new_data.loc[new_data['polarity']>0,'sentiment']= 'positive'
new_data.loc[new_data['polarity']==0,'sentiment']= 'netural'
new_data.loc[new_data['polarity']<0,'sentiment']= 'negitive'
print(new_data.head())
```
